```python
gap = pd.read_csv('gapminder.csv', index_col=0)
gap.head()
```

```python
gap[['1800','1899']].dropna().plot(kind='scatter', x='1800', y='1899')
plt.xlabel('Life Expectancy by Country in 1800')
plt.ylabel('Life Expectancy by Coyntry in 1899')
plt.xlim(20, 55)
plt.ylim(20, 55)
plt.show()
```

```python
def check_null_or_valid(row_data):
    no_na = row_data.dropna()[1:-1]
    numeric = pd.to_numeric(no_na)
    ge0 = numeric >= 0
    return ge0
```

```python
gap.columns[-1] == 'Life expectancy'
```

```python
gap.iloc[:, :-1].apply(check_null_or_valid, axis=1).all().all()
```

```python
gap[gap['Life expectancy']=='Lao']
```

```python
gap.loc[:,'1800':'1899'].dropna(how='all')
```

```python
gap.loc[:,'1900':'1999'].dropna(how='all')
```

```python
gap.loc[:,'2000':'2016'].dropna(how='all')
```

```python
gap_ = gap.groupby('Life expectancy').sum(min_count=1)
gap_.head()
```

```python
gap_.shape
```

```python
gap_melt = pd.melt(gap_.reset_index(), id_vars='Life expectancy')
gap_melt.head()
```

```python
gap_melt.columns = ['country', 'year', 'life_expectancy']
gap_melt.head()
```

```python
gap_melt['year'] = pd.to_numeric(gap_melt['year'])
gap_melt.info()
```

```python
countries = gap_melt['country']
countries = countries.drop_duplicates()
countries
```

```python
pattern = '^[A-Za-z\.\s]*$'
mask = countries.str.contains(pattern)
countries.loc[~mask]
```

```python
gap_melt_ = gap_melt.dropna()
gap_melt_.shape
```

```python
gap_melt_['life_expectancy'].plot(kind='hist')
```

```python
gap_melt_.groupby('year')['life_expectancy'].mean().plot()
plt.title('Life expectancy over the years')
plt.ylabel('Life expectancy')
plt.xlabel('Year')
```
